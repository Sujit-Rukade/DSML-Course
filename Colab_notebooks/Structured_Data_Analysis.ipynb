# Problem Statement 1

```python
# Read the file in DataFrame
df = pd.read_csv('cricket.csv')
```

```python
df
```

```python
# List the name of the cricketer and their respective runs
df.loc[:,['name','runs']]
```

```python
# Find total wickets taken by them
df['wickets'].sum()
```

```python
# Find average of the catches taken
df['catches'].mean()
```

```python
# Find the name of the wicketkeeper
x = df[df['stumpings']>0]
print("Name of the Wicketkeeper: ",x['name'])
```

```python
# Print the name of the bowler who played highest no. of matches
bowler_matches = df.groupby('name')['matches'].sum()
print("The bowler who played the highest no. of matches: ",bowler_matches.idxmax())
```

```python
# Find average of all the bowlers
df['bowling_average'] = df['runs']/df['wickets']
print("Average of all the bowlers is: \n")
df.loc[:,['name','bowling_average']]
```

```python
# Find the name of the bowler with least bowling average

min_avg = df.loc[df['bowling_average'].idxmin(), 'name']
print("Name of the bowler with the least bowling average:", min_avg)
```

```python
# Print information about all the players whose catches per match is greater than 0.5 and store it in a csv file

df['cbmr'] = df['catches']/df['matches']
greater_than_5 = df[df['cbmr']>0.5]
print(greater_than_5)
greater_than_5.to_csv('cbmr_greater_than_half.csv', index=False)
```

```python
# Draw the bar chart of player name against their runs

import matplotlib.pyplot as plt
plt.bar(df['name'], df['runs'], color= 'magenta')
plt.title("Players vs Runs")
plt.ylabel("Runs")
plt.xlabel("Name of the players")
```

```python
# Sort and print information about players by ascending order of runs

df.sort_values(by='runs',ascending=True)
```

```python
# Print the names of the players whose wickets are greater than matches

wgtm = df[df['wickets']>df['matches']]
print(wgtm['name'])
```

```python
df = pd.read_csv('fb.csv')
df.head()
```

```python
df.shape
```

```python
# Count how many number of posts are paid and unpaid

df['Paid'].value_counts()
```

```python
# Display only Page type, Month, likes, comments and shares

df.loc[:, ['Type','Post Month', 'like', 'comment', 'share']]
```

```python
# Find average number of likes to the post in September month
print("The average number of likes to the post in September month: ",df[df['Post Month']==9]['like'].mean())
```

```python
# Print the information about the posts which are links
df[df['Type']=='Link']
```

```python
# How many total shares are present for the video posts

print("Total no. of shares on video posts: ",df[df['Type']=='Video']['share'].sum())
```

```python
# In which hour of the day maximum post are posted

print(df['Post Hour'].value_counts().idxmax())
```

```python
# In which day of October, maximum no. of interactions are found?

df[df['Post Month']==10]['Post Weekday'].value_counts().idxmax()
```

```python
# Print the information of the posts where likes are above 200 and shares are above 30

df[(df['like'] > 200) & (df['share'] > 30)]
```

```python
# Which kind of posts are posted maximum no. of times in November month?

print("Kind of posts posted maximum no. of times in November: ",df[df['Post Month']==11]['Type'].value_counts().idxmax())
```

```python
# Find average and median of interactions for the paid (1) posts

print("Average: ", df[df['Paid']==1]['Total Interactions'].mean())
print("Median: ", df[df['Paid']==1]['Total Interactions'].median())
```
