# Filtering Using the Variance Threshold

```python
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(threshold = 1)
```

```python
vt.fit(df)
```

```python
newdf = df.iloc[:, vt.get_support()]
```
