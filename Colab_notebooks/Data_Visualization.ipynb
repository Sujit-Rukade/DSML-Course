```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([12,3,45,65,4,78,34,64,84,55])
y = np.array([45,65,2,7,96,45,94,47,24,85])
z = np.array([65,77,88,53,23,56,24,5,42,99])
```

# Line Plot

```python
plt.title('Line Plot of x vs y')
plt.xlabel('The value of array x')
plt.ylabel('The values in the array y')
plt.plot(x,y, marker='o', mfc= 'r', ls= '--', color = 'magenta', ms = 7, lw = 2);        # 'mfc' is marker font color, 'ls' is line style, 'ms' is marker size, 'lw' is line width
# Markers --> <, >, ^, D, o
plt.grid()        # Used for grid lines to appear
```

```python
plt.title('Line Plot of y vs z')
plt.xlabel('The value of array y')
plt.ylabel('The values in the array z')
plt.plot(y,z, marker = 'X', mfc='b', color= 'red')
```

```python
plt.title('Line Plot')
plt.plot(x,y, marker='o', mfc= 'r', ls= '--', color = 'magenta', ms = 7, lw = 2, label= '2023');        # 'mfc' is marker font color, 'ls' is line style, 'ms' is marker size, 'lw' is line width
plt.plot(y,z, marker = 'X', mfc='b', color= 'red', label= '2022');
# Markers --> <, >, ^, D, o
plt.grid();        # Used for grid lines to appear
plt.legend();
```

# Subplot

```python
plt.subplot(2,2,1);
plt.title('Line Plot of x vs y')
plt.xlabel('The value of array x')
plt.ylabel('The values in the array y')
plt.plot(x,y, marker='o', mfc= 'r', ls= '--', color = 'magenta', ms = 7, lw = 2);        # 'mfc' is marker font color, 'ls' is line space, 'ms' is marker size, 'lw' is line width
# Markers --> <, >, ^, D, o
plt.grid();        # Used for grid lines to appear

plt.subplot(2,2,2);
plt.title('Line Plot of y vs z')
plt.xlabel('The value of array y')
plt.ylabel('The values in the array z')
plt.plot(y,z, marker = 'X', mfc='b', color= 'red');
plt.grid();
```

# Bar Graph

```python
courses = ['python', 'java', 'c++', 'c', 'R']
students = [56,98,87,18,11]
```

```python
plt.title('Courses vs Students enrolled', fontsize = 15)
plt.xlabel('Courses', fontsize = 12)
plt.ylabel('No. Of Students Enrolled', fontsize = 12)
# plt.bar(courses, students, color= 'violet', width= 0.4, edgecolor= 'black', linewidth= 2, alpha= 1);       # 'alpha' is used to set the transperancy level
plt.bar(courses, students, color= ['violet', 'blue', 'red', 'yellow', 'purple'], width= 0.4, edgecolor= 'black', linewidth= 2, alpha= 1, ls = '--');
plt.grid()
```

# Scatter PLot

```python
plt.title('Scatter Plot')
plt.xlabel('The values in the array x')
plt.ylabel('The values in the array y')
plt.scatter(x,y, marker= 'o')
```

# Histogram

```python
marks = np.array([45,76,89,34,79,90])
```

```python
plt.title('Histogram')
plt.ylabel('Count')
plt.xlabel('Marks')
plt.xlim(0,100)
plt.hist(marks, bins= range(50,101,10));
```

```python
# Use subplot
# combine lineplot, bar graph, scatter plot and piechart
plt.figure(figsize= (16,10))       # Solution for the problem of overlapping of the subplots
plt.subplot(2,2,1)
plt.title('Courses vs Students enrolled', fontsize = 15)
plt.xlabel('Courses', fontsize = 12)
plt.ylabel('No. Of Students Enrolled', fontsize = 12)
# plt.bar(courses, students, color= 'violet', width= 0.4, edgecolor= 'black', linewidth= 2, alpha= 1);       # 'alpha' is used to set the transperancy level
plt.bar(courses, students, color= ['violet', 'blue', 'red', 'yellow', 'purple'], width= 0.4, edgecolor= 'black', linewidth= 2, alpha= 1, ls = '--');
plt.grid()

plt.subplot(2,2,2)
plt.title('Scatter Plot')
plt.xlabel('The values in the array x')
plt.ylabel('The values in the array y')
plt.scatter(x,y, marker= 'o')

plt.subplot(2,2,3)
plt.title('Line Plot')
plt.plot(x,y, marker='o', mfc= 'r', ls= '--', color = 'magenta', ms = 7, lw = 2, label= '2023');        # 'mfc' is marker font color, 'ls' is line style, 'ms' is marker size, 'lw' is line width
plt.plot(y,z, marker = 'X', mfc='b', color= 'red', label= '2022');
# Markers --> <, >, ^, D, o
plt.grid();        # Used for grid lines to appear
plt.legend();

plt.subplot(2,2,4)
plt.title('Pie Chart')
plt.pie(students, labels= courses, autopct= "%.2f%%", startangle= 90, radius= 0.8, shadow= True, labeldistance= 1.1)    # 'autopct' is used to mark percentage value
plt.legend()

plt.savefig('My first subplot')
# plt.tight_layout()
```

# Box Plot

```python
marks = [78,67,97,55,86,56,45,139]
plt.boxplot(marks);
```

```python
q1 = np.percentile(marks,25)
q3 = np.percentile(marks,75)
print(q1)
print(q3)
```

```python
iqr = q3-q1
iqr
```

```python
upper_bound = q3 + 1.5*iqr
lower_bound = q1 - 1.5*iqr
print(upper_bound)
print(lower_bound)
```
